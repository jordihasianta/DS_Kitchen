# HYPERPARAMETER

Pada Sebelum membahas hyperparameter, kita perlu mengetahui terlebih dahulu fungsi dari parameter dalam membentuk model machine learning.

Model parameter adalah konfigurasi dari internal model yang nilainya ditentukan oleh model itu sendiri saat training data.

Ciri-ciri parameter dalam sebuah model adalah:
1. Dibutuhkan model saat menentukan sebuah prediksi
2. Mendefinisikan kemampuan model dalam menangani masalah dalam dataset
3. Nilainya diestimasi atau dihasilkan dari dataset
4. kadang tidak ditentukan oleh pembuat model

Contoh parameter model adalah:
1. Penentuan bobot dalam Artificial Neural Network
2. Support vector dalam Support Vector Machine
3. koefisien/slope pada regresi linear atau regresi logistik

Hyperparameter adalah konfigurasi dari eksternal model yang nilainya tidak bisa diestimasi oleh model dalam training data. Hyperparameter ini biasanya ditentukan oleh pembuat model sebelum proses training data. Hyperparameter bertujuan untuk

Ciri-ciri hyperparameter adalah:
1. Seringkali digunakan pada processing untuk memperkirakan nilai parameter model.
2. Ditentukan oleh pembuat model.
3. tidak mempunyai formula yang pasti dalam menentukan hyperparameter terbaik

Umumnya, jika kita harus menentukan parameter sendiri saat membuat sebuah model, maka kemungkinan besar parameter tersebut adalah hyperparameter. Hyperparameter biasanya harus diinisialisasi sebelum melatih sebuah model.
<br>Contoh Hyperparameter:
1. learning rate pada neural network
2. Jumlah layer pada neural network
2. C dan sigma pada Support Vector Machine
3. ukuran k pada k-nearest neighbors.

Karena kita tidak mempunyai formula yang pasti untuk menentukan hyperparameter, maka yang hanya kita bisa lakukan adalah mencoba satu per satu lewat trial and error. Ada beberapa teknik untuk menentukan hyperparameter terbaik, yang kita bahas kali ini adalah RandomSearch dan GridSearch.

# Grid Search

Pada GridSearch, kita mencoba setiap kombinasi pada setiap hyperparameter dan akan dievaluasi kualitasnya. Jadi saat semua kombinasi hyperparameter telah dievaluasi, GridSearch akan memberikan kombinasi hyperparameter yang mempunyai akurasi terbaik dengan cross-validation.


Kelemahan dari GridSearch adalah dimensinya. Jadi semakin banyak dimensi parameter yang kita terapkan pada GridSearch, maka jumlah evaluasi akan semakin meningkat secara eksponensial. Hal ini menyebabkan waktu kita untuk mencari hyperparameter akan semakin lama.

# Random Search

Random Search adalah teknik menemukan hyperparameter terbaik dengan pendekatan secara acak. Kemungkinan teknik ini mencari parameter yang optimal bisa melebihi Grid Search karena random search bisa menemukan sebuah pola jika 1 parameter terbaik sudah ditemukan. Hal ini membuat Random Search memakan waktu lebih singkat secara signifikan, karena Random Search tidak perlu mengecek satu per satu untuk menemukan parameter yang optimal.

Kekurangan dari Random Search ini adalah kita tidak tahu apakah parameter yang dihasilkan adalah parameter yang terbaik atau tidak. Selain itu, kita tidak mendapatkan informasi/alasan mengapa Random Search memilih suatu kombinasi parameter.
