# CROSS VALIDATION

Sebelum memasuki tahap produksi, maka kita harus menguji kualitas sebuah model dalam tahap validasi. Pemisahan test data dan training data tidak cukup untuk menjamin seberapa baik kualitas model kita, karena bisa jadi dalam pemisahan data tersebut, persebaran test data dan training data tidak begitu mirip. Selain itu, teknik pemisahan data ini tidak menjamin keakuratan model pada dataset yang baru jika jumlah observasi pada training data sangat sedikit.

Cross Validation adalah teknik untuk menguji kualitas sebuah model. Teknik ini membagi dataset menjadi subset-subset. Jadi konsepnya sama yaitu memisah data menjadi training data dan test data, namun pada setiap iterasi dataset dipisah ke subset-subset yang baru untuk menjamin kualitas model.

Cross Validation mempunyai beberapa tipe, yaitu:
1. LOOCV - Leave One Out Cross Validation
2. K-Fold Cross validation
3. Stratified Cross Validation
4. Time Series Cross Validation

# Leave One Out Cross Validation

Pada teknik ini, kita membagi data menjadi 2 bagian. Bagian pertama kita mempunyai 1 data observasi yang akan menjadi test data, sementara bagian sisanya adalah training data. Jadi jika kita mempunyai data dengan jumlah observasi sebanyak n, maka training data akan berjumlah n-1 observasi.

Hasil dari evaluasi LOOCV akan didapat dari perhitungan rata-rata error pada tiap iterasi

Keuntungan LOOCV:
1. mempunyai nilai bias yang cukup rendah karena kita melakukan pengujian model di setiap observasi
2. Melakukan LOOCV berulang kali tetap akan menghasilkan nilai evaluasi yang sama, karena LOOCV tidak bersifat acak.

Kerugian LOOCV:
1. Nilai error akan bervariasi pada tiap iterasi. Hal ini akan menjadi lebih bermasalah jika di dalam data kita terdapat outlier
2. Tidak efisien terhadap waktu komputasi karena pengujian dilakukan sebanyak n

# K-Fold Cross Validation

Teknik ini secara acak membagi dataset menjadi k bagian. Bagian pertama digunakan untuk test data, sedangkan sisanya untuk training data. Proses ini kemudian diulang sebanyak k-1, dengan menetapkan satu bagian yang baru menjadi test data, sedangkan sisanya untuk training data.

Jadi secara intuisi prosesnya mirip dengan LOOCV. Perbedaannya adalah, sebanyak n observasi,  K-Fold Cross Validation mengelompokkan datapoint sebanyak n/k untuk dijadikan sebagai test data.

Keuntungan:
1. Secara komputasi lebih efisien dari LOOCV karena hanya diproses sebanyak k
2. mengurangi bias

Kerugian:
1. Tidak menjamin kualitas sebuah model jika dataset bersifat imbalance


# Stratified Cross Validation

Teknik ini umumnya mirip dengan K-Fold Cross Validation, bedanya adalah teknik ini mengatur sedemikian rupa agar tiap-tiap fold mampu merepresentasikan data secara keseluruhan. Jadi Stratified Cross Validation menjamin agar masing-masing fold mempunyai persentase yang sama pada tiap-tiap kelas. 

Teknik Cross Validation ini mempunyai performa yang cukup bagus dalam menangani imbalanced class.

# Time Series Cross Validation

Dalam menangani dataset yang bersifat time series, kita tidak bisa menggunakan teknik-teknik cross validation seperti diatas, karena hasil evaluasinya nanti akan menjadi tidak bisa diinterpretasikan.

Untuk time series cross validation digunakan metode rolling-origin, yang artinya teknik ini akan mengevaluasi hasil prediksi pada waktu ke n berdasarkan data pada waktu sebelumnya.

Jadi misalnya jika test data adalah waktu ke n, maka training data adalah data di waktu sebelum n.

Berdasarkan gambar diatas, kita akan memulai evaluasi model dengan jumlah observasi yang sedikit, dengan test data adalah D5. kemudian pada evaluasi berikutnya kita menetapkan D6 sebagai test data dan D1-D5 sebagai training data. Hal ini terus berlanjut sampai semua data point yang terdapat dalam dataset berhasil dimasukkan ke dalam model untuk mengetahui hasil evaluasi.
